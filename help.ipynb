→ Runs one full simulation day.

### `advance_time()`

→ Switches between day/night and increments day counter.

→ Adds NPCs into the world.

### `register_location()`

→ Adds locations into the world.

### `log_event()`

→ Records important events.

### `show_world_status()`

→ Displays current world state.

### `check_escape_condition()`

→ Checks if the mystery has been solved.

### `trigger_random_event()`
